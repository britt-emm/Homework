## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# create bins and bin labels for age column
players_bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]

players_labels = ['<10','10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
# bin the age column
age_range = pd.cut(players_df['Age'], players_bins, labels = players_labels)
# assign it as a new column in the dataframe
players_df.loc[:, 'Age Range'] = pd.Series(age_range, index=players_df.index) 
# create series that shows the count of players by age
players_age_count = pd.Series(players_df.groupby('Age Range')['Age'].count())
# remember you calculated total players earlier

# calculate percentage of players in each age range
players_age_percentage = pd.Series(players_age_count / total_players * 100)
# create a dictionary with the two series
age_demo= {'Total Count': players_age_count, 'Percentage of Players': players_age_percentage}
# concat the two series into a new dataframe
age_demo_df = pd.concat(age_demo,
                       axis =1)
# round the decimals for ease of reading
age_demo_df.round(decimals=2)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# column headers: 'Purchase Count', 'Average Purchase Price' 'Total Purchase Value', 'Avg Total Purchase per Person'
#create a new dataframe to manipulate
age_purchases_df = pd.DataFrame(purchase_data)
# create bins and bin labels for age column
players_bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]
players_labels = ['<10','10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
# bin the age column
age_range_all = pd.cut(purchase_data['Age'], players_bins, labels = players_labels)
# assign it as a new column in the dataframe
age_purchases_df.loc[:, 'Age Range'] = pd.Series(age_range_all, index=age_purchases_df.index) 
# create series for purchase count
age_purchase_count = pd.Series(age_purchases_df.groupby('Age Range')['Purchase ID'].count())
# create series for avg purchase price
age_avg_price = pd.Series(age_purchases_df.groupby('Age Range')['Price'].mean())
# create series for total purchase value
age_total_val = pd.Series(age_purchases_df.groupby('Age Range')['Price'].sum())
# create series for avg total per person
age_avg_total = pd.Series(age_total_val / players_age_count)
# create a dictionary with the series
age_analysis = {'Purchase Count': age_purchase_count,
               'Average Purchase Price(USD)': age_avg_price,
               'Total Purchase Value(USD)': age_total_val,
               'Avg Total Purchase per Person(USD)': age_avg_total}
# concat the series to create a new dataframe
age_analysis_df = pd.concat(age_analysis, axis = 1)
age_analysis_df.style.format({'Average Purchase Price(USD)': '${:.2f}', 
                              'Total Purchase Value(USD)': '${:.2f}', 
                              'Avg Total Purchase per Person(USD)': '${:.2f}'})


```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spenders_df = pd.DataFrame(purchase_data)


# create series for purchase count
spend_purchase_count = pd.Series(spenders_df.groupby('SN')['Purchase ID'].count())

# create series for avg purchase price
user_avg_price = pd.Series(spenders_df.groupby('SN')['Price'].mean())
# create series for total purchase value
user_total_val = pd.Series(spenders_df.groupby('SN')['Price'].sum())
# create a dictionary of series
user_spending = {'Purchase Count': spend_purchase_count,
                'Average Purchase Price(USD)': user_avg_price,
                'Total Purchase Value(USD)': user_total_val}
# create the new dataframe by concatenating the series
top_spenders_df = pd.concat(user_spending, axis =1)
# sort the dataframe by purchase value
top_spenders_sorted_df = top_spenders_df.sort_values('Total Purchase Value(USD)', ascending=False)
# show just the top 5 spenders
top_spenders_sorted_df.head().style.format({'Average Purchase Price(USD)': '${:.2f}', 
                              'Total Purchase Value(USD)': '${:.2f}'})
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Purchase count, Item Price, Total Purchase Value sorted by purchase count
popular_df = pd.DataFrame(purchase_data)

# create series for purchase count
pop_purchase_count = pd.Series(popular_df.groupby(['Item ID', 'Item Name'])['Purchase ID'].count())
# create series for purchase price since each item only has one price min or max should work
item_price = pd.Series(popular_df.groupby(['Item ID', 'Item Name'])['Price'].min())
# create series for total purchase value
pop_total_val = pd.Series(popular_df.groupby(['Item ID', 'Item Name'])['Price'].sum())

# create a dictionary with the series
pop_analysis = {'Purchase Count': pop_purchase_count,
               'Item Price(USD)': item_price,
               'Total Purchase Value(USD)': pop_total_val}
# concat the series to create a new dataframe
pop_analysis_df = pd.concat(pop_analysis, axis = 1)
# sort the dataframe by purchase count
pop_analysis_sorted_df = pop_analysis_df.sort_values('Purchase Count', ascending=False)
# show just the top 5 most popular
pop_analysis_sorted_df.head().style.format({'Item Price(USD)': '${:.2f}', 
                              'Total Purchase Value(USD)': '${:.2f}'})
```
