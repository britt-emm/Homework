```python
# Don't modify this cell.
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# Don't modify this cell
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city in `cities` using a series of successive API calls.
* OPTIONAL: Include a print log of each city as it'sbeing processed (with the city number and city name).


```python

# base url for getting api data
base_url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key

# example request
req = requests.get(base_url + f'&q={cities[0]}')
data = req.json()
```

```python
data["sys"]["country"]
```

```python
# example request
#req = requests.get(base_url + f'&q={cities[0]}').json()

# Your code here. A loop maybe? 
#NOTE: API calls can be slow. They can also be limited. Once you have a list try not to rerun all API calls very often.

# Create data structure to hold data points 

city_weather = {"city":[],
                "country":[],
                "humidity %":[],
                "latitude":[],
                "current temp F":[],
                "max temperature F":[],
                "wind speed mph":[],
                "cloudiness %":[]}

for city in cities:
    # base url for getting api data
    base_url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
    
    # send http request
    req1 = requests.get(base_url + f'&q={city}')
    # convert data to json
    data1 = req1.json()
    if data1["cod"] == "404":
        continue
    #append data to city_weather
    city_weather["city"].append(data1["name"])
    city_weather["country"].append(data1["sys"]["country"])
    city_weather["humidity %"].append(data1["main"]["humidity"])
    city_weather["latitude"].append(data1["coord"]["lat"])
    city_weather["current temp F"].append(data1["main"]["temp"])
    city_weather["max temperature F"].append(data1["main"]["temp_max"])
    city_weather["wind speed mph"].append(data1["wind"]["speed"])
    city_weather["cloudiness %"].append(data1["clouds"]["all"])


    
```

### Convert Raw Data to DataFrame
* OPTIONAL: Export the city data into a .csv.
* Display the DataFrame

```python
# convert to dataframe    
weather_df= pd.DataFrame.from_dict(city_weather)
weather_df.to_csv('weather_data_temp_max.csv')
print(weather_df.describe())
```

### Plotting the Data

#### Latitude vs. Temperature Plot

```python
# Title: City Latitude vs. Max Temperature (04/06/2021)
# y-axis: Max Temperature(F)
# x-axis Latitude
x = weather_df["latitude"]
y = weather_df["max temperature F"]
fig, ax = plt.subplots()
ax.scatter(x, y, color= 'm', edgecolor='black', marker='o')
ax.set(title="City Latitude vs. Max Temperature (04/06/2021)",
       ylabel="Max Temperature(F)",
       xlabel="Latitude")
plt.grid()
plt.savefig('city_weather_lat_vs_temp_plot.png')
plt.show()
```

#### Latitude vs. Humidity Plot

```python
# Title: City Latitude vs. Humidity (04/06/2021)
# y-axis: Humidity (%)
# x-axis Latitude
x = weather_df["latitude"]
y = weather_df["humidity %"]
fig, ax = plt.subplots()
ax.scatter(x, y, color= 'm', edgecolor='black', marker='o')
ax.set(title="City Latitude vs. Humidity (04/06/2021)",
       ylabel="Humidity (%)",
       xlabel="Latitude")
plt.grid()
plt.savefig('city_weather_lat_vs_humidity_plot.png')
plt.show()
```

#### Latitude vs. Cloudiness Plot

```python
# Title: City Latitude vs. Cloudiness (04/06/2021)
# y-axis: Cloudiness (%)
# x-axis Latitude
x = weather_df["latitude"]
y = weather_df["cloudiness %"]
fig, ax = plt.subplots()
ax.scatter(x, y, color= 'm', edgecolor='black', marker='o')
ax.set(title="City Latitude vs. Cloudiness (04/06/2021)",
       ylabel="Cloudiness (%)",
       xlabel="Latitude")
plt.grid()
plt.savefig('city_weather_lat_vs_cloud_plot.png')
plt.show()
```

#### Latitude vs. Wind Speed Plot

```python
# Title: City Latitude vs. Wind Speed (04/06/2021)
# y-axis: Wind Speed (mph)
# x-axis Latitude
x = weather_df["latitude"]
y = weather_df["wind speed mph"]
fig, ax = plt.subplots()
ax.scatter(x, y, color= 'm', edgecolor='black', marker='o')
ax.set(title="City Latitude vs. Wind Speed (04/06/2021)",
       ylabel="Wind Speed (mph)",
       xlabel="Latitude")
plt.grid()
plt.savefig('city_weather_lat_vs_wind_plot.png')
plt.show()
```
