```python
def dot_product(v1, v2):
    try:
        products = []
        for i in range(len(v1)):
            products.append(v1[i] * v2[i])
        return sum(products)
    except ValueError:
        print('Try transposing one of the vectors')
    except:
        print('something else went wrong')
            
    
def cross_product(v1, v2):
    try:
        cross = np.array([(v1[1]*v2[2]) - (v1[2]*v2[1]),(v1[2]*v2[0])-(v1[0]*v2[2]),(v1[0]*v2[1])-(v1[1]*v2[0])])
        return cross
    except:
        print('something went wrong')
```

### why one-hot encoding
Because one hot encoding creates a binary vector, it is useful in making sure that if the things you're encoding are of equal value the computer doesn't weigh one more than the other. So for example if male is [0,1] and female is [1,0] then as it looks at a lot of data they will still be weighted the same. 

```python
import pandas as pd
import numpy as np
```

```python
data_to_load = 'belize_airbnb_listings.csv'
```

```python
belize_df = pd.read_csv(data_to_load)
belize_df.head()
```

```python
grouped_df = pd.DataFrame(belize_df['neighbourhood_group'].value_counts())
#assign a name to the index
grouped_df.index.name = 'neighborhood'
# create a new dataframe from the old dataframe with a reset index (so index is 0,1,2,...)
new_df = grouped_df.reset_index()

neighborhood_list = new_df['neighborhood'].tolist()
neighborhood_list
```

```python
# define a mapping of groups to integers
group_to_int = dict((group, i) for i, group in enumerate(neighborhood_list))
int_to_group = dict((i, group) for i, group in enumerate(neighborhood_list))
# encode integers
integer_encoded = [group_to_int[group] for group in neighborhood_list]
# one hot encode
# create an empty list
one_hot_encoded = list()
# write a for loop that will create a list of 0s and one 1 for each group in the integer_encoded list
for value in integer_encoded:
    neighbor = [0 for _ in range(len(neighborhood_list))]
    neighbor[value] = 1
    one_hot_encoded.append(neighbor)
print(one_hot_encoded)


```
