```python
#Create a function that accepts two values, a and b. Add code to print three lines where:
# The first line contains the sum of the two numbers.
# The second line contains the differences of the two numbers (first-second)
# The third line contains the product of the two numbers
#Conditions: If any of the results are zero, print "zero" instead of 0
#be sure to include useful info pritned out, not just the result
# for example a + b = 6, a -b = zero

def homework(a, b):
    #function that manipulates a and b
    add = a + b
    subtract = a - b
    multiply = a * b
    if add == 0:
        print('A + b = zero')
    else:
        print('A + b = ' + str(add))
    if subtract == 0:
        print('A - b = zero')
    else:
        print('A - b = ' + str(subtract))
    if multiply == 0:
        print('A * b = zero')
    else:
        print('A * b = ' + str(multiply))
    
```

```python
homework(3,5)
```

```python
homework(3,3)
```
