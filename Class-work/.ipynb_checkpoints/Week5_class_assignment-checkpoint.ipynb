```python
def descending(val = 111111):
    if val > 0:
        val = str(val)
    new_val = [n for n in val]
    new_val.sort(reverse=True)
    print(''.join(new_val))
# preferable to pass a default value in case nothing is passed in        
```

```python
descending(1984)
```

```python
descending(94385398)
```

```python
descending()
```

```python
import pandas as pd
import numpy as np
import datetime

file_to_load = 'JEOPARDY_CSV.csv'

jeopardy_df = pd.read_csv(file_to_load)
jeopardy_df.head()
```

```python
jeopardy_df.describe()
```

#### There are 216930 rows and each row represents a Jeopardy question and answer from a particular show.

```python
# number of rows in dataframe. A row represents a jeopardy question from a particular show.
jeopardy_df['Show Number'].count()
```

```python
# how many shows were there

unique_shows= jeopardy_df['Show Number'].value_counts()
unique_shows
```

```python
jeopardy_df.columns= jeopardy_df.columns.str.rstrip()
jeopardy_df.columns= jeopardy_df.columns.str.lstrip()
```

```python
jeopardy_df.columns
```

```python
jeopardy_df['Air Date'].head()
```

```python
jeopardy_df['year'] = pd.DatetimeIndex(jeopardy_df['Air Date']).year
jeopardy_df.head()
jeopardy_df.tail()
```

### Unique shows in each year

```python
unique_shows_year = pd.Series(jeopardy_df.groupby('year')['Show Number'].nunique())
print(unique_shows_year)
```
