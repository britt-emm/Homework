## Write a function that lists and counts the number of divisors for an input value

```python
def get_divisors(n):
    divisors = []
    end = []
    numbers = [x for x in (range(n+1))]
    if n == 1:
        return print('There is 1 divisor: 1.')
    for number in numbers:
        if number == 0:
            continue
        if number < n:
            if n % number == 0:
                divisors.append(number)
        if number == n:
            end.append('and')
            end.append(number)
    total = len(divisors)+1 
    divisor_string = ', '.join([str(elem) for elem in divisors])
    end_string = ' '.join([str(elem) for elem in end])
    print('There are', total, 'divisors:', divisor_string, end_string)
    #return total, divisors
    
```
