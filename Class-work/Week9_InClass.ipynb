#### 1a. 
A linear correlation looks at how two variables are related in a linear fashion. Two variables with a positive linear correlation would have a positive slope on a graph, and two variable with a negative linear correlation would have a negative slope. No correlation would have a slope closer to 0. An example would be height and finishing times for marathon racers. You could see if there is a linear correlation between those two variables.

A rank correlation looks at how the rankings of two variable are related. Or looking at how two variables that are on a scale compare so looking at US Olympic medal winners and their education levels.

```python
belize_listings = 'belize_airbnb_listings.csv'
```

```python
belize_df = pd.read_csv(belize_listings)
belize_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 1b

```python
listing_avail = sns.scatterplot(x= belize_df['calculated_host_listings_count'], y = belize_df['availability_365'])
listing_avail.set_title('Relationship Between Number of Listings and Availability')
listing_avail.set(xlabel='Number of listings', ylabel='Number of days available')
plt.show()
```

```python
x = belize_df['calculated_host_listings_count']
y = belize_df['availability_365']
round(x.corr(y), 4)
```

#### 1c
The correlation coefficient is 0.1728. There is a very small positive linear correlation between the nubmer of listings and the number of days availability.

```python
fig, ax = plt.subplots()
ax.boxplot(belize_df['price'], showfliers = False)
ax.set_title('Box Plot for Price')
plt.show()
```

The box plot shows: the 25th -75th percentile is the box, the organge line is the median or the middle price of all the listed prices. The outer lines show the range of prices. However outliers can exist outside the range because the box plot shows a distribution of the data. This box plot excludes outliers.
