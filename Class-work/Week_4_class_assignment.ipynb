```python
og_list = [2,4,6,8,10,12,14]
squares = [x ** 2 for x in og_list]
print(squares)
```

```python
two_thirds = [x * 2/3 for x in og_list]
print(two_thirds)
```

```python
dict={"Apples": 1500, "Cranberries": 2000, "Pears": 2500, "Cherries": 1600, "Plums": 2400, "Peaches": 13600, "Oranges": 7}
# prints a new version of the dictionary with all the keys lowered
lowercase = {keys.lower():value for (keys,value) in dict.items()}
print(lowercase)
# prints a list of just the keys lowercased
lower = [keys.lower() for keys in dict]
print(lower)
```
