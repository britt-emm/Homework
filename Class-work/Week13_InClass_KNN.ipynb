### 1. Write a function that predicts the ideal K value
It should iterate over a range of values and return the k and teh score when the accuracy score is maximized

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

diabetes_df = pd.read_csv("../week_13/diabetes.csv")
diabetes_df.head()
    
```

```python
X = diabetes_df.drop('Outcome', axis=1)
y = diabetes_df['Outcome']

#split data into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=45)

#Standardization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)

knn = KNeighborsClassifier(n_neighbors=5)

# fitting on the training data
knn.fit(X_train,y_train)
#y_predicted = knn.predict(X_test)
#print(y_predicted)
```

```python
# testing on the test data
# verifying my accuracy score
print(knn.score(X_test,y_test))
```

### Here is my function

```python
def get_K(xvalue, yvalue):
    '''Calculating the KNN value that gives the highest accuracy score'''
    accuracies = []
    values = []
    for i in (range(20)):
        if i == 0:
            continue
        knn = KNeighborsClassifier(n_neighbors=i)
        # fit the knn value to the training values
        knn.fit(X_train, y_train)
        accuracies.append(knn.score(xvalue, yvalue))
        values.append(i)
        accuracy_value = zip(accuracies, values)
    # this will print the accuracy,value tuple (created by zipping the accuracies and values above) with the highest
    # accuracy value
    print(max(accuracy_value))
```

```python
get_K(X_test,y_test)
```
