### Write a function to concatenate the data back into sentences with one readable sentence in each list.

```python
import nltk

sentences = nltk.corpus.gutenberg.sents("carroll-alice.txt")[:10]
print(sentences)
```

```python
# if you want the sentences to still have some punctuation, but some awkward spaces are acceptable

def new_sentence(text):
    for sentence in text:
        new_list = [' '.join(x for x in sentence)]
        filtered_list = [x.replace(' .', '.') for x in new_list]
        filtered_list2 = [x.replace(' ,', ',') for x in filtered_list]
        filtered_list3 = [x.replace(' ?', '?') for x in filtered_list2]
        filtered_list4 = [x.replace(' !', '!') for x in filtered_list3]
        filtered_list5 = [x.replace(' \' ', '\'') for x in filtered_list4]
    
        print(filtered_list5)
    
    
    
```

```python
# if you don't mind if the sentences dont' have punctuation at all
def new_sentence2(text):
    for sentence in text:
        filtered_list = [word for word in sentence if word.isalnum()]
        print([' '.join(x for x in filtered_list)])
```

Both functions present readable sentences, even though they aren't grammatically correct.

```python
new_sentence(sentences)
```

```python
new_sentence2(sentences)
```

## Create a function that accepts a list of values. Filter out all values that are numeric and return the new list.

```python
my_list = [1,2,'a','b',7.6]

def numeric(some_list):
    new_list = []
    for x in some_list:
        if str(x).isalpha() == True:
            new_list.append(x)
    return new_list
```

```python
numeric(my_list)
```

```python
your_list = [4, '5', 'hat', 5.7, 'b']
numeric(your_list)
```
