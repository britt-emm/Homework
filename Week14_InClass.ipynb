```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_selection import RFE
from sklearn.svm import SVR

crime_df = pd.read_csv('crime_data.csv')
```

### Question 1 Apply Recursive Feature Selection to the crime dataset

Recursive Feature Selection or Recursive Feature Elimination (RFE) assigns weights to features and then selects the features that seem to be the most important and then continues to eliminate features until the optimal number of features is selected. Recursion means the function calls itself.

```python
crime_df.info()
```

```python
X = crime_df.drop('X3', axis =1)
y = crime_df['X3']

estimator = SVR(kernel ='linear')
selector =RFE(estimator, n_features_to_select=3)
selector = selector.fit(X,y)
print(selector.support_)
print(selector.ranking_)
```

### Question 2 Create a function called digital_root that takes in an integer. Digital root is the recursive sum of all the digits in a number.

```python
"""how to break up an integer into individual digits, convert it to a string
and then convert the string back to an integer"""
n = 16
digit_list = [int(d) for d in str(n)]
digit_list
```

```python
# double checking that you can sum the contents of a list
n = [2,3,4,1]
sum(n)
```

```python
def digital_root(n):
    '''calculates the recursive sum of all the digits in a number'''
    digit_list = [int(d) for d in str(n)]
    new = sum(digit_list)
    if new < 10:
        return new
    else:
        digit2 = [int(d) for d in str(new)]
        next = sum(digit2)
        if next < 10:
            return next
        else:
            digit3 = [int(d) for d in str(next)]
            third = sum(digit3)
            return third
    
        
```
